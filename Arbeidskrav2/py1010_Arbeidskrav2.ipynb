# Oppg 3) 
Lag et program med en funksjon som regner om fra grader til radianer.
Programmet skal starte med:
import numpy as np
v_grad = float(input('Skriv inn gradtallet:' ))
Radiantallet til vinkelen regnes så ut ved følgende formel: v_rad = v_grad*np.pi/180
Resultatet v_rad skrives til skjerm med passende tekst og verdi.
Merk: np.pi er en ferdiglaget funksjon som gir verdien 3.1415....

```python
"""
Created on Wednesday Jan 15 2025

@author: henninggundersen

Dette er et Python-progam som leveres inn til til oppgave 3 i Arbeidskrav2 i PY101-1 24H
program skal lages med en funksjon som regner om fra grader til radianer.
Jeg har valgt å presentere svaret med 2 desimaler

"""
import numpy as np

# funksjon som konverterer antall grader til radianer
def konverterGraderTilRadianer(v_grad):
    v_rad = v_grad*np.pi/180
    return v_rad
    
# Programmet ber om at bruker legger inn gradtallet som skal regnes om til radianer
v_grad = float(input('Skriv inn gradtallet:' ))

# skriver ut hvor mange grader og hva dette er i radianer
print(str(v_grad) + ' grader er omregnet til ' + str("%.2f" % konverterGraderTilRadianer(v_grad)) + ' radianer')
```

# Oppg 5) 
Lag et program med en funksjon som tar a og b som inn-argumenter og som så
regner ut arealet og «ytre» omkrets til en figur satt sammen av en rettvinklet trekant og en
halvsirkel, se figuren under. Med «ytre» omkrets menes samlet lengde av de sorte strekene.
Funksjonen skal returnere arealet og «ytre» omkrets, som så skrives til skjerm med passende
tekst.



```python
"""
Created on Wednesday Jan 16 2025

@author: henninggundersen

Dette er et Python-progam som leveres inn til til oppgave 5, i Arbeidskrav2 i PY101-1 24H
Lag program med en funksjon som tar a og b som inn-argumenter og som så
regner ut arealet og «ytre» omkrets til en figur satt sammen av en rettvinklet trekant og en
halvsirkel, som vist i oppgave

"""

import numpy as np
import math

# funksjon som tar imot 2 argumenter (diameter sirkel og trekant lengde), og regner ut omkrets og areal av figuren
def beregnOmkretsOgAreal(a, b):
    radiusSirkel = a/2
    lengdeTrekantKortside = a
    lengdeTrekantLangside = b
    
    #beregne areal av en halv sirkel
    arealSirkelDel = (np.pi * radiusSirkel**2) / 2
    
    #beregne omkrets av en halv sirkel
    omkretsSirkelDel = 2 * np.pi * radiusSirkel / 2

    #beregne areal av trekant
    arealTrekant = lengdeTrekantKortside * lengdeTrekantLangside / 2
    
    #beregne omkrets av en trekant (minus den kortsiden som ligger mot sirkel)
    omkretsTrekant = math.sqrt(lengdeTrekantKortside**2 + lengdeTrekantLangside**2) + lengdeTrekantLangside 

    #returnerer samlet areal og omkrets
    return [arealSirkelDel + arealTrekant, omkretsSirkelDel + omkretsTrekant]

# test av funksjon
diameterSirkel = 4
lengdeTrekant = 8
figurData = beregnOmkretsOgAreal(diameterSirkel,lengdeTrekant)


# skrive dataene på figuren basert på test data
print('figuren med verdiene a=' + str(diameterSirkel) + 'og b=' + str(lengdeTrekant) + ' har følgende data (med 2 desimaler): ' )
print('totalt areal : ' + str("%.2f" % figurData[0]))
print('total ytre omkrets : ' + str("%.2f" % figurData[1]))
```

# Oppg 6) 
Skriv en kode som plotter funksjonen 𝑓(𝑥) = −𝑥2 − 5, for x på intervallet [-10,10].
Hint: np.linspace(-10, 10, 200) gir en array med 200 punkter jevnt fordelt på intervallet
[-10,10].

```python
"""
Created on Wednesday Jan 16 2025

@author: henninggundersen

Dette er et Python-progam som leveres inn til til oppgave 6, i Arbeidskrav2 i PY101-1 24H
Skriv en kode som plotter funksjonen 𝑓(𝑥) = −𝑥2 − 5, for x på intervallet [-10,10].


"""

import numpy as np
import matplotlib.pyplot as plt

# definerer intervallet for x med antall punkter = 200
x = np.linspace(-10, 10, 200)

# definerer funksjonen for y verdier f(x)
yFunction = -(x**2)-5

# tegner plot
plt.plot(x, yFunction)

# setter på grid for lesbarhet
plt.grid()

# setter Label på X-akse
plt.xlabel("X")

# setter Label på X-akse
plt.ylabel("f(x)")

# setter på overskrift over plot
plt.title("oppgave 6, i Arbeidskrav2 i PY101-1 24H")

#fjerner text over plot
plt.show()
```
